### task_#3

```python
x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax = ax.flatten()

ax[0].plot(x, y1, color = '#008B8B')
ax[1].plot(x, y2, color = 'orange')
ax[2].plot(x, y3, color = '#DC143C')
ax[3].plot(x, y4, color = 'green')

ax[0].set_xlim([-5, 5])
fig.set_size_inches(8, 6)

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
cc_df = pd.read_csv('creditcard.csv')

#pd.value_counts(cc_df['Class']).plot(kind = 'bar')
pd.value_counts(cc_df['Class']).plot(kind = 'bar', logy = True)
plt.show()
```

```python
data_1 = cc_df.loc[(cc_df['Class'] == 1) , 'V1']
data_0 = cc_df.loc[(cc_df['Class'] == 0) , 'V1']
data_1.sort_values(inplace = True)
data_0.sort_values(inplace = True)
```

```python
plt.hist(data_1, color = 'red',  bins = 20, alpha = 0.5, label = 'Class 1', density = True, ec = 'black')
plt.hist(data_0, color = 'grey', bins = 20, alpha = 0.5, label = 'Class 0', density = True, ec = 'blue')
plt.xlabel('V1')
plt.ylabel('density')
plt.axis([-30.5, 2, 0, 0.3])
plt.legend(loc = 'upper left')
plt.show()
```

### task_13

```python
D_inv[D_inv < 0] = 0
D_inv[D_inv > 0] = 1
log_0 = D_inv == 0
log_1 = D_inv == 1
assert np.all(log_0 + log_1)
E = np.where(D_inv, B, C)
assert (np.all(E[log_1] == B[log_1]) & np.all(E[log_0] == C[log_0]))
```
