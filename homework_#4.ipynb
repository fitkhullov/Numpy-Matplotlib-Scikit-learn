```python
D_inv[D_inv < 0] = 0
D_inv[D_inv > 0] = 1
log_0 = D_inv == 0
log_1 = D_inv == 1
assert np.all(log_0 + log_1)
E = np.where(D_inv, B, C)
assert (np.all(E[log_1] == B[log_1]) & np.all(E[log_0] == C[log_0]))
```
